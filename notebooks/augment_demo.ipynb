# 🧪 Disaster Vision Augmentation Demo
Visualize how real-world UAV noise and variation is simulated.

```python
import cv2
import matplotlib.pyplot as plt
from augmentor.augment import *

img = cv2.imread('../data/sample.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
# Apply all augmentations and visualize
augs = [flip_horizontal, rotate_random, add_gaussian_noise, gaussian_blur, brightness_contrast_shift, random_crop]
fig, axs = plt.subplots(2, 3, figsize=(15, 10))
for i, aug in enumerate(augs):
    ax = axs[i // 3, i % 3]
    aug_img = aug(img)
    ax.imshow(aug_img)
    ax.set_title(aug.__name__)
    ax.axis('off')
plt.tight_layout()
plt.show()
```
